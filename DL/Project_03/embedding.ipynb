```python
import numpy as np

import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
sent=['the glass of milk', 'the glass of juice', 'the cup of tea',
    'I am a good boy', 'I am a good developer', 'understand the meaning of words', 'your blogs are good',]
```

```python
voc_size = 10000 # 
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
# Padding docs to create equal length vectors
sent_len = 8
padded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_len)
padded_docs
```

```python
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_len))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(padded_docs) # Each word will be represented by vector of length 10 as set in dim
```

```python
model.predict(padded_docs[0])
```

### 2. Modelling

```python
from tensorflow.keras.datasets import imdb
```

```python
max_features = 10000 # vocabulary size
(Xtrain, ytrain), (Xtest, ytest) = imdb.load_data(num_words=max_features)
```

```python
print(f'Train size - {Xtrain.shape}, Train labels shape - {ytrain.shape}')
print(f'Test size - {Xtest.shape}, Test labels shape - {ytest.shape}')
```

```python
len(Xtrain), len(Xtrain[0]), len(Xtrain[1]) # Thus, one hot encoded vectors of each sentences
```

```python
sample_review, sample_label = Xtrain[0], ytrain[0]
print(sample_review, '\n', sample_label)
```

```python
## Mapping of words to indexes
word_index = imdb.get_word_index()
word_index
```

```python
word_index_reversed = {value:key for key, value in word_index.items()}
word_index_reversed[1]
```

```python
decoded_review = ' '.join([word_index_reversed.get(i-3, '?') for i in sample_review])
decoded_review
```

```python
max_len = 500
Xtrain = pad_sequences(Xtrain, maxlen=max_len, padding='pre')
Xtest = pad_sequences(Xtest, maxlen=max_len, padding='pre')
```

```python
Xtrain[0]
```

```python
## RNN Architecture
model = Sequential()
model.add(Embedding(max_features, 128, input_length=max_len))
model.add(SimpleRNN(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
## Create an instance of EarlyStopping Callback
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
early_stopping
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(Xtrain, ytrain, epochs=10, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

```python
model.save('rnn_imdb.h5')
```
